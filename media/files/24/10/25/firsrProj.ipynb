```python
import cv2
import numpy as np
import os

# تحميل الصورة
img = cv2.imread('sss.png')

# تحديد أحجام النوافذ
window_sizes = [3, 5, 9, 15]

# تحديد أبعاد الصورة
height, width, channels = img.shape

# تحديد نوع البيانات الناتجة
dtype = np.float32

# إنشاء مجلد لحفظ الصور (إذا لم يكن موجودًا)
# if not os.path.exists('output_images'):
#     os.makedirs('output_images')

# حلقات لتطبيق الفلاتر وحفظ الصور
for window_size in window_sizes:
    # حساب نصف حجم النافذة
    half_window = window_size // 2

    # إنشاء مصفوفات لنتائج المتوسط والوسيط
    mean_img = np.zeros_like(img, dtype=dtype)
    median_img = np.zeros_like(img, dtype=dtype)

    for i in range(height):
        for j in range(width):
            # تحديد حدود النافذة
            row_start = max(i - half_window, 0)
            row_end = min(i + half_window + 1, height)
            col_start = max(j - half_window, 0)
            col_end = min(j + half_window + 1, width)

            # استخراج المجاورة
            neighborhood = img[row_start:row_end, col_start:col_end, :]

            # حساب المتوسط يدويًا
            mean_value = np.sum(neighborhood) / (window_size * window_size * channels)
            mean_img[i, j] = mean_value

            # حساب الوسيط يدويًا (طريقة بسيطة، قد لا تكون الأفضل من حيث الكفاءة)
            sorted_neighborhood = np.sort(neighborhood, axis=None)
            index = sorted_neighborhood.size // 2
            median_value = sorted_neighborhood[index]
            median_img[i, j] = median_value

    # حفظ الصور
    cv2.imwrite(f'images3/mean_{window_size}x{window_size}.png', mean_img.astype(np.uint8))
    cv2.imwrite(f'images3/median_{window_size}x{window_size}.png', median_img.astype(np.uint8))
```

```python
from PIL import Image
import numpy as np
import os

def calculate_mean_median(image_path, filter_size):
    # Load the image
    image = Image.open(image_path)
    pixels = np.array(image)

    # Ensure filter size is odd and in range
    filter_size = max(3, min(filter_size, 15))
    if filter_size % 2 == 0:
        filter_size += 1

    # Padding for border handling
    pad_size = filter_size // 2
    padded_image = np.pad(pixels, ((pad_size, pad_size), (pad_size, pad_size), (0, 0)), mode='constant', constant_values=0)

    # Initialize result arrays
    mean_image = np.zeros_like(pixels, dtype=float)
    median_image = np.zeros_like(pixels, dtype=float)

    # Iterate over each pixel in the image
    for i in range(pixels.shape[0]):
        for j in range(pixels.shape[1]):
            # Extract the neighborhood
            neighborhood = padded_image[i:i + filter_size, j:j + filter_size, :]

            # Calculate mean and median for each color channel
            for channel in range(3):  # R, G, B
                flattened_neighborhood = neighborhood[:, :, channel].flatten()
                mean_image[i, j, channel] = np.mean(flattened_neighborhood)
                median_image[i, j, channel] = np.median(flattened_neighborhood)

    return mean_image.astype(np.uint8), median_image.astype(np.uint8)

def save_images(image_path):
    # Create output directory if it does not exist
    if not os.path.exists('output_images'):
        os.makedirs('output_images')
    
    # Loop over filter sizes from 3 to 15
    ba=[3,5,9,15]
    for filter_size in ba:
        mean, median = calculate_mean_median(image_path, filter_size)

        # Convert results to images
        mean_image = Image.fromarray(mean)
        median_image = Image.fromarray(median)

        # Save images with filter size in the name
        mean_image.save(f"output_images/image_mean_{filter_size}.jpg")
        median_image.save(f"output_images/image_median_{filter_size}.jpg")

# Example usage
save_images("R.jpeg")

```

```python
import cv2
import numpy as np
import os

# Load the image
image = cv2.imread('m.jpg')

# Function to calculate the average of a 3x3 filter for all color channels
def calculate_average_3x3(img, i, j):
    total = np.sum(img[i-1:i+2, j-1:j+2], axis=(0,1))
    return total // 9

# Function to calculate the median of a 3x3 filter for all color channels
def calculate_median_3x3(img, i, j):
    values = np.sort(img[i-1:i+2, j-1:j+2].reshape(-1, 3), axis=0)
    return values[4]

# Function to apply filter on image for all color channels
def apply_filter(img, filter_size, filter_func):
    filtered_image = np.zeros_like(img)
    padding = filter_size // 2

    for i in range(padding, img.shape[0] - padding):
        for j in range(padding, img.shape[1] - padding):
            filtered_image[i, j] = filter_func(img, i, j)

    return filtered_image

# Apply filters for all color channels and save filtered images
filter_sizes = [3, 5, 9, 15]
filter_functions = [calculate_average_3x3, calculate_median_3x3]

for filter_size in filter_sizes:
    for func in filter_functions:
        filtered_image = np.zeros_like(image)
        for channel in range(3):
            filtered_channel = apply_filter(image[:, :, channel], filter_size, func)
            filtered_image[:, :, channel] = filtered_channel

        cv2.imwrite(f'filtered_{func.__name__}_{filter_size}.jpg', filtered_image)

# Create a directory to save filtered images
if not os.path.exists('filtered_images'):
    os.makedirs('filtered_images')
    for filename in os.listdir():
        if filename.startswith('filtered_'):
            os.rename(filename, os.path.join('filtered_images', filename))

print("Filtered images saved successfully!")
```
