```python
import cv2

# تحميل classifiers
face = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
eye = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_eye.xml')
mouth = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_mcs_mouth.xml')
profile_face = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_profileface.xml')
upper_body = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_upperbody.xml')
full_body = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_fullbody.xml')
lower_body = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_eye_tree_eyeglasses.xml')
smile = cv2.CascadeClassifier('C:/Users/ComputerWorld/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_smile.xml')

# قراءة الصورة
img = cv2.imread('m.jpg')
img = cv2.resize(img, (500, 500))
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

img2 = cv2.imread('man_picture.jpg')
img2 = cv2.resize(img2, (500, 500))
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

img3 = cv2.imread('uu8.jpg')
img3 = cv2.resize(img3, (600, 600))
img_gray3 = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)

# الكشف عن الوجوه
faces = face.detectMultiScale(img_gray, 1.3, 5)
for (x, y, w, h) in faces:
    img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = img_gray[y:y + h, x:x + w]
    roi_color = img[y:y + h, x:x + w]
    # الكشف عن العينين
    eyes = eye.detectMultiScale(roi_gray, 1.3, 5)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 1)

    # الكشف عن الشفاه
    mouths = mouth.detectMultiScale(roi_gray, 1.3, 5)
    for (mx, my, mw, mh) in mouths:
        if my > h // 2:  # تأكد من أنه تحت العينين
            cv2.rectangle(roi_color, (mx, my), (mx + mw, my + mh), (200, 200, 255), 1)
            
    # الكشف عن الابتسامة
    smiles = smile.detectMultiScale(roi_gray, 1.3, 5)
    for (sx, sy, sw, sh) in smiles:
        cv2.rectangle(roi_color, (sx, sy), (sx + sw, sy + sh), (200, 100, 200), 1)
    
# الكشف عن الجزء العلوي من الجسم
upper_bodies = upper_body.detectMultiScale(img_gray2, 1.3, 5)
for (ux, uy, uw, uh) in upper_bodies:
    cv2.rectangle(img2, (ux, uy), (ux + uw, uy + uh), (255, 255, 255), 2)
    
faces2 = face.detectMultiScale(img_gray3, 1.3, 5)
for (x, y, w, h) in faces2:
    img3 = cv2.rectangle(img3, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray2 = img_gray3[y:y + h, x:x + w]
    roi_color2 = img3[y:y + h, x:x + w]
    # الكشف عن الجزء السفلي من الجسم
    lower_bodies =lower_body.detectMultiScale(roi_color2, scaleFactor=1.3, minNeighbors=5)
    for (ux, uy, uw, uh) in lower_bodies:
        cv2.rectangle(img3, (ux+10, uy+40), (ux + uw+35, uy + uh+45), (100, 200, 100), 4)


# عرض الصورة
# cv2.imshow('image1', img)
# cv2.imshow('image2', img2)
cv2.imshow('image3', img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
